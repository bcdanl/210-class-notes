# Classwork 7 - Q7
Calculate a â€˜Normalized_Sugarsâ€™ variable for each product by Manufacturer, where the normalization formula is

$$\text{Normalized_Sugars}
=
\frac{Sugars
âˆ’
mean(Sugars)}
{std(Sugars)}$$

for each Manufacturer group. This formula adjusts the sugar content of each product by subtracting the mean sugar content of its manufacturer and then dividing by the standard deviation of the sugar content within its manufacturer.



```python
cereal.shape[0]
```

```python
cereal_mf[['Sugars']].transform('mean')
```

```python
cereal = (
    cereal
    .assign(
        sugar_mean = cereal_mf[['Sugars']].transform('mean'),
        sugar_std = cereal_mf[['Sugars']].transform('std')
    )
)

cereal['Normalized_Sugars'] = ( cereal['Sugars'] - cereal['sugar_mean'] ) / cereal['sugar_std']

cereal
```

```python
(
    cereal
    .assign(
        Normalized_Sugars2 = ( cereal['Sugars'] - cereal_mf['Sugars'].transform('mean') )
                                / cereal_mf['Sugars'].transform('std')
    )
)
```
