```python
import pandas as pd
# Below is for an interactive display of DataFrame in Colab
from google.colab import data_table
data_table.enable_dataframe_formatter()

fortune1000 = pd.read_csv("https://bcdanl.github.io/data/fortune_2023.csv")

varlist = ['Company', 'Revenues_M', 'Profits_M', 'Number_of_employees', 'Sector', 'Industry']
fortune = fortune1000[varlist]
```

```python
fortune_shuffled = (
    fortune
    .sample( frac = 1 )
    .reset_index( drop = True )
)
```

```python
fortune_shuffled = (
    fortune
    .sample( frac = 1 )
    .reset_index(drop = True)
)
fortune_shuffled
```

```python
(
    fortune_shuffled
    .groupby("Sector")
    .sort_values("Revenues_M", ascending = False)
)
```

```python
(
    fortune_shuffled
    .groupby("Sector")
    .nlargest(5, "Revenues_M", keep= 'all')
)
```

```python
def get_largest_obs(df):
  return df.nlargest(1, "Revenues_M", keep = 'all')
```

```python
(
    fortune_shuffled
    .groupby("Sector")
    .apply(get_largest_obs)
)
```

```python
(
    fortune_shuffled
    .sort_values(['Revenues_M'], ascending = False)
    .groupby('Sector')
    .first() # or nth(0)
)
```

```python
(
    fortune_shuffled
    .groupby('Sector')
    .apply(lambda df: df.nlargest(1, "Revenues_M", keep = 'all'))
)
```

```python
def get_largest_obs_2(df, n, var):
  return df.nlargest(n, var, keep='all')
```

```python
(
    fortune_shuffled
    .groupby('Sector')
    .apply(get_largest_obs_2)
)
```

```python
(
    fortune_shuffled
    .groupby('Sector')
    .apply(get_largest_obs_2, 5, "Revenues_M")
)
```

## Q2

- For each market, find the average price_per_floz, total dollar_spent, and total beer_floz.
- Which market has the highest average price_per_floz?

```python
df = (
    beer
    .groupby('market')
    .agg(
        price_per_floz_mean = ('price_per_floz', 'mean'),
        dollar_spent_total = ('dollar_spent', 'sum'),
        beer_floz_total = ('beer_floz', 'sum')
    )
    .sort_values('price_per_floz_mean', ascending = False)
)

df
```
