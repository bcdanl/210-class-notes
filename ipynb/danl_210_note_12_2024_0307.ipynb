# Q8

Identify the days when Netflix added only one movie to its catalog.



```python
netflix['type'].value_counts(dropna=False)
```

```python
netflix[ (netflix['type'] == "Movie") | (netflix['date_added'].notna())  ]
```

```python
(
    netflix
    .astype({'date_added':'datetime64'})[ (netflix['type'] == "Movie") & (netflix['date_added'].notna())  ]
    .drop_duplicates(subset = ['date_added'], keep = False)[['date_added']]
    .value_counts(dropna=False)
    .shape
)
```
