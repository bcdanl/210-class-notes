# assign() method

```python
import pandas as pd
# Below is for an interactive display of DataFrame in Colab
from google.colab import data_table
data_table.enable_dataframe_formatter()
```

```python
df = pd.DataFrame({'A': range(1, 6), 'B': range(10, 15)})
df
```

```python
df['C'] = df['A'] + df['B']
df
```

```python
df = df.drop(columns = 'C')
```

```python
df
```

```python
df = df.assign(
    C = df['A'] + df['B']
)
df
```

```python
(
    df.assign(
    D = df['A'] + df['B'] + df['C'],
    E = df['A'] * df['B'])
    .describe()

)

```

```python
(
    df.assign(
    D = df['A'] + df['B'] + df['C'],
    E = df['A'] * df['B'],
    F = df['E'] + df['E']
)

)
```

```python
(
    df
    .assign(
        C = lambda x: x['A'] + x['B'],
        D = lambda x: x['A'] * x['B']
    )
    .query('C > 15')
)
```

```python
# Creating a dictionary
food_data = {
"Item": ["Apple", "Onion", "Orange", "Tomato", "Watermelon"],
"Type": ["Fruit", "Vegie", "Fruit", "Vegie", "Fruit"],
"Price": [1.05, 1.00, 1.25, 0.85, 4.15]
}

# Converting a dictionary into a DataFrame
supermarket = pd.DataFrame(data = food_data)
supermarket
```

```python
groups = supermarket.groupby("Type")
```

```python
groups
```

```python
groups['Price'].mean()
```
